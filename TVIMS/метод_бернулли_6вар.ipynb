```python
import math

# ============================================================
# ЗАДАЧА 1
# ============================================================
print("ЗАДАЧА 1")
n = 100000
m = 5800
p = 0.06
q = 1 - p

# Используем логарифмы, чтобы избежать переполнения
logP = (math.log(math.comb(n, m)) +
        m * math.log(p) +
        (n - m) * math.log(q))
P = math.exp(logP)

print(f"Решение: P = C({n},{m}) · {p}^{m} · {q}^{n-m}")
print(f"P = {P:.6e}")
print()

# ============================================================
# ЗАДАЧА 2
# ============================================================
print("ЗАДАЧА 2")
n = 1200
p = 0.002
m = 0

# При m=0 формула упрощается: P = q^n
logP = m * math.log(p) + (n - m) * math.log(1 - p)
P = math.exp(logP)

print(f"Решение: P = C({n},{m}) · {p}^{m} · (1-{p})^{n-m} = (1-{p})^{n}")
print(f"P = {P:.6f}")
print()

# ============================================================
# ЗАДАЧА 3
# ============================================================
print("ЗАДАЧА 3")
n = 1000
p = 0.8
m = 1000

# При m=n формула упрощается: P = p^n
logP = m * math.log(p)
P = math.exp(logP)

print(f"Решение: P = {p}^{n}")
print(f"P = {P:.6e}")
print()

# ============================================================
# ЗАДАЧА 4
# ============================================================
print("ЗАДАЧА 4")
n = 30
p = 0.75
m = 30

logP = m * math.log(p)
P = math.exp(logP)

print(f"Решение: P = {p}^{n}")
print(f"P = {P:.6e}")
print()

# ============================================================
# ЗАДАЧА 5
# ============================================================
print("ЗАДАЧА 5")
n = 50
p = 0.88
m = 50

logP = m * math.log(p)
P = math.exp(logP)

print(f"Решение: P = {p}^{n}")
print(f"P = {P:.6f}")
print()

# ============================================================
# ЗАДАЧА 6
# ============================================================
print("ЗАДАЧА 6")
n = 80
p = 0.7
m = 80

logP = m * math.log(p)
P = math.exp(logP)

print(f"Решение: P = {p}^{n}")
print(f"P = {P:.6e}")
print()

# ============================================================
# ЗАДАЧА 7 (придуманная)
# ============================================================
print("ЗАДАЧА 7")
print("Дано: 40 выстрелов, вероятность попадания = 0.95")
print("Найти: вероятность 40 попаданий")
n = 40
p = 0.95
m = 40

logP = m * math.log(p)
P = math.exp(logP)

print(f"Решение: P = {p}^{n}")
print(f"P = {P:.6f}")
print()
```
